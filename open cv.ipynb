```python
from skimage import data, color
import matplotlib.pyplot as plt

# Load the rocket image
rocket_image = data.rocket()

# Convert the image to grayscale
gray_image = color.rgb2gray(rocket_image)

# Convert the grayscale image back to RGB using the original RGB image
rgb_image = color.gray2rgb(gray_image)

# Display the grayscale image
plt.imshow(gray_image, cmap='gray')
plt.title('Grayscale Image')
plt.axis('off')
plt.show()


plt.imshow(rocket_image)
plt.title('RGB Image')
plt.show()


```

# flip images 

```python
import cv2
import matplotlib.pyplot as plt

def show_image(image, title=''):
    plt.imshow(image, cmap='gray')
    plt.title(title)
    plt.axis('off')
    plt.show()
```

```python
import numpy as np
# Flip the image vertically
seville_vertical_flip = np.flipud(gray_image)

```

```python
# Flip the image horizontally
seville_horizontal_flip = np.flipud(gray_image)

# Show the resulting image
show_image(seville_horizontal_flip, 'Seville')
```

# Histograms

```python
# Obtain the red channel
red_channel = rocket_image[:, :, 0]

# Plot the red histogram with bins in a range of 256
plt.hist(red_channel.ravel(), bins=256)

# Set title and show
plt.title('Red Histogram')
plt.show()
```
